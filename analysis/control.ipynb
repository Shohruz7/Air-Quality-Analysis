```python
import sys, certifi, requests, pandas as pd
print("sys.executable:", sys.executable)   # should show /usr/local/bin/python3

# quick SSL + data test
url = "https://data.cityofnewyork.us/resource/c3uy-2p5r.json?$limit=5"
resp = requests.get(url, verify=certifi.where(), timeout=20)
resp.raise_for_status()
df = pd.DataFrame.from_records(resp.json())
print(df.shape)
df.head()
```

```python
# pip install requests certifi
import io
import pandas as pd
import requests
import certifi

url = 'https://data.cityofnewyork.us/resource/c3uy-2p5r.json?$limit=50000'

resp = requests.get(url, verify=certifi.where())
resp.raise_for_status()
df = pd.read_json(io.StringIO(resp.text))
print(df.shape)
df.head()
```
